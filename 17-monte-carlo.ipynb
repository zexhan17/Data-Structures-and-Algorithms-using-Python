# Estimating Pi

Area of circle is given as: πr^2





So, π is essentially a ratio between the area and the radius square. We can use that to estimate the value of π using Monte Carlo simulation.

We can simulate throwing darts and see which ones land within the circle and which one land outisde it.

area(circle) = πr^2

area(square) = (2r)^2 = 4r^2

So, if we divide the area of circle by the area of square, we get π/4. Multiply by 4 and you get an estimate of the value of π.

```python
import random as r
from math import sqrt
```

```python
def distance(a, b):
    return sqrt(a**2 + b**2)

def estimate_pi(darts):
    inside = 0
    
    for i in range(darts):
        x = r.random()
        y = r.random()
        
        if distance(x, y) < 1.0:
            inside += 1
        
    pi = (inside / darts) * 4
    return pi
```

```python
num_sim = int( 1e+5)
pi = estimate_pi(num_sim)
print(pi)
```

# Random Walks

```python
steps = 200
```

```python
import random

def get_action():
    return random.choice( [ (0, 1), (0, -1), (1, 0), (-1, 0) ] )

def random_walk(n):
    """Return path after 'n' block"""
    
    path = []
    
    x, y = 0, 0
    for i in range(n):
        
        (dx, dy) = get_action() # get step deltas
        
        x = x+dx
        y = y+dy
        
        path.append( (x, y) ) # create new tuple and save in the 'path'
    
    return path

walk = random_walk(steps)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure( num=None, figsize=(15, 10))

point_color = list(range(steps))

x_numbers = [x[0] for x in walk]
y_numbers = [x[1] for x in walk]

plt.scatter(x_numbers, y_numbers, marker='.', s=100, c=point_color)

plt.show()
```
